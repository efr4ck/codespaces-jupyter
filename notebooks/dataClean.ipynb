```python
import pandas as pd
```

```python
df = pd.read_csv("../dataPrisma/test.csv")
df
```

```python
finalDF = df[['Jobs.JobName', 'PrintActions.Destination', 'Jobs.PrismaJobID','PrintActions.OriginalPages','PrintActions.Copies', 'PrintActions.TimeStart', 'PrintActions.TimeEnd']]
finalDF
```

```python
finalDF['dateTS'] = finalDF['PrintActions.TimeStart'].str.extract(r'(\w+\s\d+,\s\d+)')
finalDF['timeTS'] = finalDF['PrintActions.TimeStart'].str.extract(r'(\d+:\d+\s[AP]M)')
```

```python
finalDF
```

```python
finalDF['dateTE'] = finalDF['PrintActions.TimeEnd'].str.extract(r'(\w+\s\d+,\s\d+)')
finalDF['timeTE'] = finalDF['PrintActions.TimeEnd'].str.extract(r'(\d+:\d+\s[AP]M)')
```

```python
finalDF
```

```python
#TimeStart converted to datetime

finalDF['dateTS'] = pd.to_datetime(finalDF['dateTS'])

# Combine date and time into a single datetime column
finalDF['datetimeTS'] = finalDF['dateTS'].dt.strftime('%Y-%m-%d') + ' ' + finalDF['timeTS']

# Convert the combined column to datetime format
finalDF['datetimeTS'] = pd.to_datetime(finalDF['datetimeTS'])

finalDF
```

```python
#TimeEnd converted to datetime

finalDF['dateTE'] = pd.to_datetime(finalDF['dateTE'])

# Combine date and time into a single datetime column
finalDF['datetimeTE'] = finalDF['dateTE'].dt.strftime('%Y-%m-%d') + ' ' + finalDF['timeTE']

# Convert the combined column to datetime format
finalDF['datetimeTE'] = pd.to_datetime(finalDF['datetimeTE'])

finalDF
```

```python

# Calculate the time difference
finalDF['TimeDifference'] = (finalDF['datetimeTE'] - finalDF['datetimeTS']).dt.total_seconds() / 60

finalDF
```

```python
del finalDF['PrintActions.TimeStart']
del finalDF['PrintActions.TimeEnd']
del finalDF['dateTS']
del finalDF['timeTS']
del finalDF['dateTE']
del finalDF['timeTE']
```

```python

dropCopieslow = finalDF.copy()
```

```python
dropCopieslow = dropCopieslow[dropCopieslow['PrintActions.Copies'] >= 40]
dropCopieslow
```

```python
from datetime import datetime

# Get the current date and time
current_date_time = datetime.now()

# Extract the date from the datetime object
current_date = current_date_time.date()

# Print the current date
print("Current date:", current_date) 
```

```python
#Export dataframe to CSV
dropCopieslow.to_csv(f'../dataPrisma/{current_date}_reportCopies.csv', index=False)
print('Data exported succesfully')
```
